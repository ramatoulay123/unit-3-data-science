# Titanic Population 

The titanic was a ship that sailed in 1911 which contained thousands of passengers and crew members on board. The british luxury sank on april 15, 1912 taking the lives of Hundreds on board. 

```python
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
titanic = pd.read_csv("titanic_passengers.csv")    
```

# In total how many people died on the titanic?  

```python
survived = titanic['Survived'].value_counts()

# Set the background color to pink for the plot
plt.figure(facecolor='pink')

# Plotting the pie chart
survived.plot.pie(labels=["Non- Survivors", "Survived"], autopct='%1.1f%%', colors=['#ff6666', '#66b3ff'])

# Adding a title to the plot
plt.title("Survivors vs Non-Survivors")

# Display the plot
plt.show()

```

This pie chart shows that a lot of people did not survive during the titanic with 61.6% dying while 38.4% survived the titanic 

# Which place did most passengers embark from? 

```python
embark = titanic['Embarked'].value_counts()

# Display the counts in the console
print(embark)


embark.plot(kind='bar', color=['#ff9999','#ff9999' , '#ff9999'])

# Adding a title and labels
plt.title("Embark Distribution")
plt.xlabel("Embarked")
plt.ylabel("Number of Passengers")

# Show the plot
plt.show()
```

this shows that most of the passenges embarked from s with more than 600 of them, while a lot less people embarked from q with less than 100 of the passengers. 

# Did gender affect the survival Rate ?

```python
gender = titanic.groupby(["Sex"])["Survived"].sum() 
print(gender) 



# Determines color and bar graph
gender.plot(kind='bar', color=['#ff9999','#ff9999'])

# Adding a title and labels
plt.title("Survival rate based on gender")
plt.xlabel("Sex") 
plt.ylabel("Number of Passengers")

# Show the plot
plt.show()
```

This graph shows that females where more likely to survive the titanic compared to men which was less likely. 

# Did Sibsp affect the fare of an individual? 

```python
affect = titanic.groupby(["SibSp"])["Fare"].value_counts() 
print(affect)
# Plotting for sibsp and the fare using scatter plot 
plt.figure(figsize=(10, 5), facecolor="pink")
plt.scatter(x='SibSp', y='Fare', data=titanic)
plt.title('Scatter plot of Fare vs SibSp')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Fare')
plt.show() 
```

Thid graph shows that infact the number of siblings and spouses on board did not really affect the fare that people were paying at all

# How did having siblings, spouses, parents, children, and also your age affect ticket fare? 

```python
grouped1 = titanic.groupby(["Age"])["Fare"].value_counts() 

print(grouped1) 

grouped2 = titanic.groupby(["SibSp"])["Fare"].value_counts() 
print(grouped2) 

grouped3 = titanic.groupby(["Parch"])["Fare"].value_counts()
print(grouped3) 

# Plotting for the age versus the fare 
plt.figure(figsize=(5, 5), facecolor="pink")
plt.scatter(x='Age', y='Fare', data=titanic)
plt.title('Scatter plot of Fare vs Age')
plt.xlabel('Age of passengers ')
plt.ylabel('Fare')
plt.show() 
# Plotting for the fare and the sibsp 
plt.figure(figsize=(5, 5), facecolor="pink")
plt.scatter(x='SibSp', y='Fare', data=titanic)
plt.title(' Fare vs SibSp')
plt.xlabel("Sibling and spouses on board")
plt.ylabel('Fare')
plt.show() 
# Plotting for the parch and the fare 
plt.figure(figsize=(5, 5), facecolor="pink")
plt.scatter(x='Parch', y='Fare', data=titanic)
plt.title(' Fare vs Parch')
plt.xlabel("Parch")
plt.ylabel('Fare')
plt.show()  

```

If you have siblings and spouses on board the more you had the more your price decreased. As well as if you had parents or children on board the more that you had the more your fare decreases as well. For the ages the prices varied greatly and this might be based on the passenger class.
