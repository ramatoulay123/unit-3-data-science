#### 1.5 - Backgrounds

```python
plt.figure(figsize=(10, 6), facecolor='lightgray')
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.title('Gray Figure Background')
plt.show()
```

##### Question 11
What changed when we added `facecolor='lightgray'` to `plt.figure`?
- **Answer:** What changed was the color Light gray was added to the outside of the plot where the x and y values as well as the title of the plot is. 

##### Question 12
What didnt change in this example? Hyposthesize why the whole graph did not change color.
- **Answer:** The whole graph did not change color because it only affects the figure background not the actually plotting part. This also makes it better to read the data compared to if it was colored. 

Now let's change the plot area (inner) background

```python
plt.figure(figsize=(10, 6))
ax = plt.gca()  # get current axes
ax.set_facecolor('lightyellow')
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.title('Yellow Plot Area Background')
```

##### Question 13
What's the difference between `plt.figure(facecolor='...')` and `ax.set_facecolor('...')`?
- **Answer:** The difference is that plt.figure(facelor= ) only changes the color of the outside part where the title and the x and y values are rather than inside where the line is plotted, while ax. set_facecolor changes the color inside the graph where the points are graphed. 

##### Question 14
Why have two seperate functions for setting the background? Create a scenario in which that would be useful.
- **Answer:** Having separate functions for setting the figure and axes backgrounds allows for greater customization, as well as it makes reading the data better compared to if everything was the same color.

#### 2.0 - Line Plot
Let's start with a simple line plot.

```python
month = [1, 2, 3, 4, 5, 6]
sales = [2, 4, 5, 8, 9, 12]

plt.figure(figsize=(10, 6))
plt.plot(month, sales)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales (thousands)')
plt.grid(True)
plt.show()
```

#### 2.1 - Scatter Plot
Let's look at some data about students' study time and test scores:

```python
study_hours = [1, 2, 2, 3, 4, 4, 5, 5, 6]
test_scores = [65, 70, 75, 80, 85, 85, 90, 95, 100]

plt.figure(figsize=(10, 6))
plt.scatter(study_hours, test_scores)
plt.title('Study Time vs. Test Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.show()
```

#### 2.2 - Bar Plot
Let's look at the favorite pets in a classroom:

```python
categories = ['Dogs', 'Cats', 'Fish', 'Birds']
values = [12, 8, 3, 2]

plt.figure(figsize=(10, 6))
plt.bar(categories, values)
plt.title('Favorite Pets in Class')
plt.xlabel('Pet Type')
plt.ylabel('Number of Students')
plt.show()
```

#### 2.3 - Histogram
Let's look at the distribution of heights in a class:

```python
heights = [65, 68, 67, 67, 70, 65, 68, 69, 66, 71, 
           67, 68, 68, 69, 70, 67, 66, 67, 68, 69]

plt.figure(figsize=(10, 6))
plt.hist(heights, bins=6, edgecolor='black')
plt.title('Student Heights')
plt.xlabel('Height (inches)')
plt.ylabel('Number of Students')
plt.show()
```
